**Завдання 1. Підрахунок загальної кількості респондентів**

```python
# кількість унікальних респондентів
respondents_count = df['ResponseId'].nunique()
print("Загальна кількість респондентів:", respondents_count)
```

**Завдання 2. Аналіз повноти відповідей респондентів**

```python
# 1. Отримуємо множину унікальних назв питань зі схеми
qnames = set(df1['qname'].dropna().unique())

# 2. Перетин множини qnames з колонками датафрейму відповідей
questions = set(df.columns).intersection(qnames)

print("Кількість питань у схемі:", len(qnames))
print("Кількість питань у перетині:", len(questions))

# 3. Видаляємо респондентів із пропусками у цих питаннях
complete_df = df.dropna(subset=questions)

# 4. Підраховуємо кількість таких респондентів
complete_count = complete_df.shape[0]

print("Кількість респондентів, які відповіли на всі запитання:", complete_count)
```

**Завдання 3. Статистичний аналіз досвіду респондентів**

```python
# Перетворення колонки WorkExp у числовий формат (бо там можуть бути рядки)
##df['WorkExp'] = pd.to_numeric(df['WorkExp'], errors='coerce')

# Обчислення показників
mean_val = df['WorkExp'].mean().round(1)
median_val = df['WorkExp'].median()
mode_val = df['WorkExp'].mode()[0]   

# Формування результату у вигляді таблиці
result = pd.DataFrame({
    'Mean': [mean_val],
    'Median': [median_val],
    'Mode': [mode_val]
})

print(result)
```

**Завдання 4. Аналіз віддаленої роботи**

```python
# Підрахунок кількості респондентів, які працюють віддалено
remote_count = (df['RemoteWork'] == 'Remote').sum()

print("Кількість респондентів, які працюють віддалено:", remote_count)
```

**Завдання 5. Визначення популярності Python**

```python
# Створюємо новий стовпець worked_with_python (True/False)
df['worked_with_python'] = df['LanguageHaveWorkedWith'].str.contains('python', case=False, na=False)

# Підрахунок кількості респондентів, які працювали з Python
python_count = df['worked_with_python'].sum()

# Загальна кількість респондентів
total_count = df.shape[0]

# Обчислення відсотка
python_percent = (python_count / total_count) * 100

print(f"Відсоток респондентів, які програмують на Python: {python_percent:.1f}%")
```

**Завдання 6. Аналіз шляхів навчання програмуванню**

```python
# Створюємо булевий стовпець: True якщо респондент згадував онлайн-курси
df['Learned_with_online_courses'] = df['LearnCode'].str.contains('online course', case=False, na=False)

# Підрахунок кількості респондентів, які навчалися через онлайн-курси
online_courses_count = df['Learned_with_online_courses'].sum()

print("Кількість респондентів, які навчалися програмуванню через онлайн-курси:", online_courses_count)
```

**Завдання 7. Географічний аналіз компенсації Python-розробників**

```python
# 1. Вибираємо лише Python-розробників (булевий стовпець ми створювали раніше)
df['worked_with_python'] = df['LanguageHaveWorkedWith'].str.contains('python', case=False, na=False)

# 2. Фільтруємо Python-розробників та видаляємо пропуски у компенсації
py_devs = df[df['worked_with_python']].dropna(subset=['ConvertedCompYearly'])

# 3. Групуємо по країнах і обчислюємо середню та медіанну компенсацію
py_progress_by_country = py_devs.groupby('Country')['ConvertedCompYearly'].agg(
    mean_compensation=lambda x: round(x.mean(), 1),
    median_compensation='median'
).reset_index()

# 4. Виводимо результати
print(py_progress_by_country.head(10))  # приклад перших 10 рядків
```

**Завдання 8. Аналіз освіти найбільш оплачуваних спеціалістів**

```python
# Вибираємо потрібні стовпці
df_subset = df[['ResponseId', 'EdLevel', 'ConvertedCompYearly']]

# Сортуємо за компенсацією у спадному порядку
df_sorted = df_subset.sort_values(by='ConvertedCompYearly', ascending=False).reset_index(drop=True)

# Вибираємо топ-5 респондентів
top5 = df_sorted.head(5)

print(top5[['EdLevel']])
```

**Завдання 9. Аналіз популярності Python по віковим категоріям**

```python
# Створюємо булевий стовпець: True якщо респондент згадував Python
df['worked_with_python'] = df['LanguageHaveWorkedWith'].str.contains('python', case=False, na=False)

py_by_age = df.groupby('Age')['worked_with_python'].mean().reset_index()
py_by_age['Python_percent'] = (py_by_age['worked_with_python'] * 100).round(1)
py_by_age = py_by_age.drop(columns='worked_with_python')

# Виводимо результати
print(py_by_age)
```

**Завдання 10. Аналіз індустрій серед високооплачуваних віддалених працівників**

```python
# 1. Розрахунок 75-го перцентиля компенсації
q75 = df['ConvertedCompYearly'].quantile(0.75)

# 2. Фільтрація: респонденти з компенсацією > 75-го перцентиля та які працюють віддалено
high_paid_remote = df[(df['ConvertedCompYearly'] > q75) & (df['RemoteWork'] == 'Remote')]

# 3. Підрахунок найпопулярніших індустрій
industry_counts = high_paid_remote['Industry'].value_counts().reset_index()
industry_counts.columns = ['Industry', 'Count']

# 4. Виведення результатів
print("75-й перцентиль компенсації:", q75, "\n")
print(industry_counts.head(10))  # топ-10 індустрій
```
